```python
import pickle
```

```python
pickle_in = open("/content/drive/MyDrive/Training Data/Training Data1(cleaned)/X (1).pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("/content/drive/MyDrive/Training Data/Training Data1(cleaned)/y (1).pickle","rb")
y = pickle.load(pickle_in)
```

```python
y
```

```python
from keras import utils
y=utils.to_categorical(y)
```

```python
y.shape
```

```python
import keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
```

```python
X=X/255.0
```

```python
model=Sequential()
model.add(Conv2D(64,(3,3),input_shape=X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))

model.add(Dense(4))
model.add(Activation("softmax"))

model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist=model.fit(X,y,batch_size=32,epochs=20,validation_split=0.1)
```

```python
pickle_in = open("/content/drive/MyDrive/TestData/test0/Xtest.pickle","rb")
Xtest = pickle.load(pickle_in)

pickle_in = open("/content/drive/MyDrive/TestData/test0/ytest.pickle","rb")
ytest = pickle.load(pickle_in)
```

```python
Xtest=Xtest/255.0
ytest=utils.to_categorical(ytest)
```

```python
ypred=model.predict(Xtest)
```

```python
(val_loss,val_accuracy)=model.evaluate(Xtest,ytest,batch_size=32,verbose=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
def model_predict(imgp,model):
  IMG_SIZE=256
  img_array = cv2.imread(imgp ,cv2.IMREAD_GRAYSCALE)  # convert to array
  new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
  arr= np.array(new_array).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
  arr=arr/255.0
  pre=model.predict(arr)
  preds='Probabilities of Classes:'
  preds=preds+'\n'+'Aluminium: '+str(pre[0][0])
  preds=preds+'\n'+'Copper: '+str(pre[0][1])
  preds=preds+'\n'+'Iron: '+str(pre[0][2])
  preds=preds+'\n'+'Manganese'+str(pre[0][3])
  return preds
```
